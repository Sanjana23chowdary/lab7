```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

data = pd.read_csv("NID_data.csv")
class1 = "normal"
class2 = "anomaly"
selected_data = data[(data['class'] == class1) | (data['class'] == class2)]
selected_data = pd.get_dummies(selected_data, columns=['protocol_type', 'service', 'flag'])

X = selected_data.drop(columns=['class'])
y = selected_data['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

svm_model = SVC(kernel='linear', C=1.0)  
svm_model.fit(X_train, y_train)

accuracy = svm_model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```
